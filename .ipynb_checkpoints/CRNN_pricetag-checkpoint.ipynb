```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,precision_score, recall_score

from CRNNmodel import CRNN
from TagDataset import TagDataset
from torch.utils.data import  DataLoader
from train import train_model, eval_model, predict_one_model

device = 'cpu'#'mps' if torch.backends.mps.is_available() else 'cpu'
```

```python
# Загружаем данные
train_dataset = TagDataset(root_dir='ocr_data/train.csv', images_dir = 'ocr_data/imgs/')
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_dataset = TagDataset(root_dir='ocr_data/val.csv', images_dir = 'ocr_data/imgs/')
val_loader = DataLoader(train_dataset, batch_size=64, shuffle=False)
```

```python
from torchvision import transforms as T
T.ToPILImage()(next(iter(train_loader))[0][0])
```

## Обучение модели
CTCLoss не реализован на mps, поэтому увы CPU

```python
# Инициализация и обучение модели
print(f"DEVICE is {device}")
model = CRNN(num_classes=11, pretrained = True,freeze_grads = True).to(device)
train_model(model, train_loader, num_epochs = 15, device = device, PATH_TO_SAVE = 'checkpoints/CRNN_pretrained.pth')
```

## Валидация и расчет метрик

```python
import torch
model = torch.load('checkpoints/CRNN_pretrained.pth',weights_only=False)
results_val = eval_model(model, val_loader, device = device)
```

Рассчет метрик

```python
print(f"accuracy = {accuracy_score(results_val['actual'], results_val['prediction'])}")
print(f"precision_macrio = {precision_score(results_val['actual'], results_val['prediction'], average = 'macro')}")
print(f"recall_macro{recall_score(results_val['actual'], results_val['prediction'], average = 'macro')}")
```

## Блок с тестовой выборкой

```python
results_test = pd.read_csv('ocr_data/test.csv')
results_test['image_path'] = 'ocr_data/imgs/' + results_test['img_name']
```

```python
results_test['predict'] = results_test.apply(lambda x: predict_one_model(model, x.image_path,device), axis=1)
```
